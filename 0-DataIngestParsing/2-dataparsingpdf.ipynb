### Handling PDF Challenges 
🎯 Purpose of This Section
PDFs are notoriously difficult to parse because they:

- Store text in complex ways (not just simple text)
- Can have formatting issues
- May contain scanned images (requiring OCR)
- Often have extraction artifacts


```python
# Example of raw PDF extraction
raw_pdf_text = """Company Financial Report


    The ﬁnancial performance for ﬁscal year 2024
    shows signiﬁcant growth in proﬁtability.
    
    
    
    Revenue increased by 25%.
    
The company's efﬁciency improved due to workﬂow
optimization.


Page 1 of 10
"""

# Apply the cleaning function
def clean_text(text):
    # Remove excessive whitespace
    text = " ".join(text.split())
    
    # Fix ligatures
    text = text.replace("ﬁ", "fi")
    text = text.replace("ﬂ", "fl")
    
    return text

cleaned = clean_text(raw_pdf_text)
print("BEFORE:")
print(repr(raw_pdf_text))
print("\nAFTER:")
print(repr(cleaned))

# Output:
# BEFORE:
# 'Company Financial Report\n\n\n    The ﬁnancial performance for ﬁscal year 2024\n    shows signiﬁcan'
# 
# AFTER:
# 'Company Financial Report The financial performance for fiscal year 2024 shows significant growth in'
```

```python
# from langchain_core.documents import Document
# from typing import List
# class SmartPDFProcessor:
#     """Advanced PDF processing with error handling"""
#     def __init__(self,chunk_size=1000,chunk_overlap=100):
#         self.chunk_size=chunk_size,
#         self.chunk_overlap=chunk_overlap,
#         self.text_splitter=RecursiveCharacterTextSplitter(
#             chunk_size=chunk_size,
#             chunk_overlap=chunk_overlap,
#             separators=[" "],

#         )

#     def process_pdf(self,pdf_path:str)->List[Document]:
#         """Process PDF with smart chunking and metadata enhancement"""

#         # Laod PDF

#         loader=PyPDFLoader(pdf_path)
#         pages=loader.load()

#         ## Process each page

#         processed_chunks=[]

#         for page_num,page in enumerate(pages):
#             ## clean text
#             cleaned_text=self._clean_text(page.page_content)

#             # Skip nearly empty pages
#             if len(cleaned_text.strip()) < 50:
#                 continue

#             # Create chunks with enhanced metadata
#             chunks = self.text_splitter.create_documents(
#                 texts=[cleaned_text],
#                 metadatas=[{
#                     **page.metadata,
#                     "page": page_num + 1,
#                     "total_pages": len(pages),
#                     "chunk_method": "smart_pdf_processor",
#                     "char_count": len(cleaned_text)
#                 }]
#             )
            
#             processed_chunks.extend(chunks)

#         return processed_chunks

#     def _clean_text(self, text: str) -> str:
#         """Clean extracted text"""
#         # Remove excessive whitespace
#         text = " ".join(text.split())
        
#         # Fix common PDF extraction issues
#         text = text.replace("ﬁ", "fi")
#         text = text.replace("ﬂ", "fl")
        
#         return text



import re
import hashlib
from typing import List
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document

class SmartPDFProcessor:
    """
    Robust PDF processor for Bangla / bilingual government documents.
    Handles:
    - Bangla-safe extraction
    - Layout preservation
    - Paragraph de-duplication
    - Retrieval-friendly chunking
    """

    def __init__(self, chunk_size=900, chunk_overlap=120):
        # IMPORTANT: no trailing commas
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap

        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap=chunk_overlap,
            separators=[
                "\n\n",   # paragraphs
                "\n",     # lines
                "।",      # Bangla sentence end
                ".", "?", "!",
                " ", ""
            ],
        )

    def process_pdf(self, pdf_path: str) -> List[Document]:
        loader = PyMuPDFLoader(pdf_path)
        pages = loader.load()

        all_chunks = []

        for page_num, page in enumerate(pages):
            cleaned_text = self._clean_text(page.page_content)

            # Skip junk pages
            if len(cleaned_text.strip()) < 80:
                continue

            chunks = self.text_splitter.create_documents(
                texts=[cleaned_text],
                metadatas=[{
                    **page.metadata,
                    "page": page_num + 1,
                    "total_pages": len(pages),
                    "chunk_method": "smart_pdf_processor",
                    "char_count": len(cleaned_text),
                }]
            )

            all_chunks.extend(chunks)

        # FINAL SAFETY: remove duplicate chunks
        unique_chunks = self._deduplicate_chunks(all_chunks)

        return unique_chunks

    # -------------------------------
    # Cleaning
    # -------------------------------
    def _clean_text(self, text: str) -> str:
        # Normalize whitespace (keep structure)
        text = re.sub(r"[ \t]+", " ", text)
        text = re.sub(r"\n{3,}", "\n\n", text)

        # Fix common ligatures
        text = text.replace("ﬁ", "fi").replace("ﬂ", "fl")

        # Remove duplicated paragraphs (VERY IMPORTANT for govt PDFs)
        paragraphs = [p.strip() for p in text.split("\n\n") if p.strip()]

        seen = set()
        deduped = []
        for p in paragraphs:
            key = re.sub(r"\s+", " ", p)
            if key not in seen:
                seen.add(key)
                deduped.append(p)

        return "\n\n".join(deduped)

    # -------------------------------
    # Chunk-level de-duplication
    # -------------------------------
    def _deduplicate_chunks(self, docs: List[Document]) -> List[Document]:
        seen = set()
        unique_docs = []

        for d in docs:
            fingerprint = self._fingerprint(d.page_content)
            if fingerprint in seen:
                continue
            seen.add(fingerprint)
            unique_docs.append(d)

        return unique_docs

    def _fingerprint(self, text: str) -> str:
        norm = re.sub(r"\s+", " ", text).strip()
        return hashlib.md5(norm.encode("utf-8")).hexdigest()

    
            
```

```python
processor = SmartPDFProcessor(chunk_size=900, chunk_overlap=120)
docs = processor.process_pdf("data/pdf/Passport.pdf")

print("Total chunks:", len(docs))

for i in range(min(3, len(docs))):
    print("\n--- Chunk", i, "---")
    print("Page:", docs[i].metadata.get("page"))
    print(docs[i].page_content[:600])
```
